## WHAT PRODUCTS ARE PARTICULARLY SOLD TOGETHER ?

```python
df
```

```python
df2 = df[df['Order ID'].duplicated(keep=False)]
df2.head(10)
```

```python
df2['Order ID'].duplicated(keep=False)
```

```python
df2['Grouped products'] = df2.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df2[['Order ID','Grouped products']].drop_duplicates()
df2.head(50)
```

```python
from itertools import combinations 
from collections import Counter 
```

```python
count = Counter()

for row in df2['Grouped products']:
    row_list = row.split(',')
    count.update(combinations(row_list,2))

for key , value in count.most_common(10):
    print(key,value)
```
